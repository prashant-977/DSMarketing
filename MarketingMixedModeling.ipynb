## CREATING A CUSTOMER JOURNEY MAP

We are trying to tackle following things:
- Persona Development: Create detailed profiles for each customer persona, including their goals, challenges, and preferences.
- Touchpoint Identification: List all possible interactions these personas might have with ZaraTech across different channels.
- Journey Mapping: Create a journey map for each persona. Include stages such as ‘Awareness’, ‘Consideration’, ‘Purchase’, and ‘Loyalty.’ Plot touchpoints and potential emotions or pain points at each stage.
- Analysis: Identify key moments of truth and pain points for each persona.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "Customer_Journey_Map_Data.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# Summarize the dataset
print(df.describe(include='all'))
```

```python
# Count of emotions per persona
emotion_count = df.groupby(['Persona', 'Emotion']).size().unstack()
print(emotion_count)
```

```python
# Count of pain points per persona
pain_point_count = df.groupby(['Persona', 'Pain Point']).size().unstack()
print(pain_point_count)
```

```python
# Plotting emotion count
sns.barplot(data=emotion_count)
plt.title("Emotion Count per Persona")
plt.ylabel("Count")
plt.show()
```

```python
# Plotting pain point count sns.barplot(data=pain_point_count)
plt.figure(figsize=(12, 6))
sns.barplot(data=pain_point_count)
plt.title("Pain Point Count per Persona")
plt.ylabel("Count")
plt.show()
```

## Excersice 2: Touchpoint Effectiveness Analysis

```python
touchpoint_file_path = "Touchpoint_Effectiveness_Analysis_Data.csv"
touchpoint_df = pd.read_csv(touchpoint_file_path)
touchpoint_df.head()
```

```python
touchpoint_df.describe()
```

 It shows that the average customer satisfaction score is about 82.6, with a mean conversion rate of 6.22% and an average of 33% for repeat visits/purchases.

```python
for column in touchpoint_df.columns:
  if column != 'Touchpoint':
    plt.figure(figsize=(10, 6))
    # Ensure unique touchpoints by setting hue='Touchpoint'
    sns.barplot(x='Touchpoint', y=column, data=touchpoint_df, hue='Touchpoint')
    plt.title(f'{column} by Touchpoint')
    plt.xlabel('Touchpoint')
    plt.ylabel(column)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

1. **Customer Satisfaction Scores by Touchpoint**: The bar plot shows the customer satisfaction scores for each touchpoint. 'In-store Visits' have the highest satisfaction score, indicating a strong performance in this area.

2. **Conversion Rates by Touchpoint**: This plot visualizes the conversion rates associated with each touchpoint. 'Email Open Rates' stand out with the highest conversion rate, suggesting that email marketing is particularly effective for this synthetic dataset.

3. **Repeat Visits/Purchases by Touchpoint**: The final plot shows the percentage of repeat visits or purchases for each touchpoint. 'In-store Visits' again show a strong performance, indicating that customers who visit the store are more likely to return or make repeat purchases.
